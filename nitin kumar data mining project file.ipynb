```python
# finding outliers in dataset
plt.figure(figsize = (12,8))
feature_list = new_pca.columns
for i in range(len(feature_list)):
    plt.subplot(3, 4, i + 1)
    sns.boxplot(y = new_pca[feature_list[i]], data = new_pca)
    plt.title('Boxplot of {}'.format(feature_list[i]))
    plt.tight_layout()
```

```python
# detect outliers
def remove_outlier(col):
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range 
```

```python
# remove outliers
for i in feature_list:
    LL, UL = remove_outlier(new_pca[i])
    new_pca[i] = np.where(new_pca[i] > UL, UL, new_pca[i])
    new_pca[i] = np.where(new_pca[i] < LL, LL, new_pca[i])
```

```python
# removed all outliers in dataset
plt.figure(figsize = (12,8))
feature_list = new_pca.columns
for i in range(len(feature_list)):
    plt.subplot(3, 4, i + 1)
    sns.boxplot(y = new_pca[feature_list[i]], data = new_pca)
    plt.title('Boxplot of {}'.format(feature_list[i]))
    plt.tight_layout()
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler()
scaled_df = X.fit_transform(new_df)
scaled_df
```

```python
from sklearn.cluster import KMeans
```

```python
wss =[] 
for i in range(1,11):
    KM =KMeans(n_clusters=i)
    KM.fit(scaled_df)
    wss.append(KM.inertia_)
```

```python
wss
```

```python
plt.plot(range(1,11), wss)
```

```python
k_means = KMeans(n_clusters = 3)
k_means.fit(scaled_df)
labels = k_means.labels_
```

```python
df_clu["Clus"] = labels
df_clu.head(10)
```
